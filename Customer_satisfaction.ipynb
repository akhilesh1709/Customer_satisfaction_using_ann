```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("customer_train.csv")
```

```python
df.head()
```

```python
df.drop(["Unnamed: 0","id"],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

Handling missing values

```python
df.isnull().sum()
```

```python
df['Arrival Delay in Minutes']=df['Arrival Delay in Minutes'].fillna(df['Arrival Delay in Minutes'].mode()[0])
```

```python
df.isna().sum()
```

Handling categorical data

```python
cols = ['Gender', 'Customer Type', 'Type of Travel', 'Class', 'satisfaction']

for col in cols:
    print(f"{col}:\n{df[col].value_counts()}\n")
```

```python
# Initialize the LabelEncoder
label_encoder = LabelEncoder()
```

```python
# Apply label encoding to each categorical column
categorical_columns = ['Gender', 'Customer Type', 'Type of Travel', 'Class', 'satisfaction']
df[categorical_columns] = df[categorical_columns].apply(lambda col: label_encoder.fit_transform(col))
```

```python
df.head()
```

Train Test split

```python
X = df.drop('satisfaction',axis='columns')
y = df['satisfaction']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
len(X_train.columns)
```

Training the model

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.callbacks import EarlyStopping
```

```python
# Example model with dropout and early stopping
model = Sequential([
    Dense(22, input_shape=(22,), activation='relu'),
    Dense(15, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
opt = keras.optimizers.Adam(learning_rate=0.01)
```

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=100)
```

```python
model.evaluate(X_test, y_test)
```

```python
yp = model.predict(X_test)
yp[:5]
```

```python
y_pred = []
for element in yp:
    if element > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:10]
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test,y_pred))
```
